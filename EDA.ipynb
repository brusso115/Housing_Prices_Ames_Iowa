# Missing Values

```python
def plotMissing(data):
    na_counts = data.isnull().sum()
    na_counts = na_counts[na_counts != 0]
    na_counts = na_counts.sort_values(ascending=False)
    plt.rcParams["figure.figsize"] = (7,7)
    sns.barplot(x=na_counts,y=na_counts.index, orient='h').set_title('Train Set Missing Values')
    plt.xlabel('Number of Missing Values')
```

# Correlation b/w Variables

```python
def get_redundant_pairs(df):
    '''Get diagonal and lower triangular pairs of correlation matrix'''
    pairs_to_drop = set()
    cols = df.columns
    for i in range(0, df.shape[1]):
        for j in range(0, i+1):
            pairs_to_drop.add((cols[i], cols[j]))
    return pairs_to_drop

def get_top_abs_correlations(df, n=5):
    au_corr = df.corr().abs().unstack()
    labels_to_drop = get_redundant_pairs(df)
    au_corr = au_corr.drop(labels=labels_to_drop).sort_values(ascending=False)
    return au_corr[0:n]
```

```python
X=train.loc[:, train.columns != 'SalePrice']
non_dummy_cols = [col for col in X if 'Dummy' not in col]
custom_features = ['TotalSF', 'BsmtBaths', 'GrLivAreaHighQualitySF', 'TotalIndoorSF','Baths','NumFloors']
X = X[non_dummy_cols]
X = X.drop(custom_features, axis=1)
```

```python
corr = X.corr()
```

```python
print("Top Absolute Correlations")
mostCorrelatedFeat = get_top_abs_correlations(corr, 1000)
mostCorrelatedFeat.head(50)
```

# Value Counts

```python
def getValueCounts(cols, data):
    f, axes = plt.subplots(9, 5, figsize=(10,10))
    rowIdx = 0
    colIdx = 0
    for col in cols:
        if colIdx > 4:
            rowIdx = rowIdx + 1
            colIdx = 0
        value_counts = data[col].value_counts()
        ax = sns.barplot(x=value_counts.index, y=value_counts, data=data, ax=axes[rowIdx,colIdx]).set_title(col)
        colIdx += 1
        
    f.tight_layout()
```

```python
cat_cols = train_org.columns[train_org.dtypes == 'object'] | train_org.columns[train_org.columns == 'MSSubClass']
getValueCounts(cat_cols, train_org)
```

# Boxplots

```python
def getBoxplots(cols, data):
    f, axes = plt.subplots(round(len(cols)/2)+1, 2, figsize=(10,80))
    rowIdx = 0
    colIdx = 0
    for col in cols:
        if colIdx > 1:
            rowIdx = rowIdx + 1
            colIdx = 0
        #print(rowIdx,colIdx)
        if col == 'GarageYrBlt':
            continue
        ax = sns.boxplot(x=col, y='SalePrice', data=data, ax=axes[rowIdx,colIdx]).set_title(col)
        data[col].value_counts()
        colIdx += 1
    f.tight_layout()
```

# Histograms

```python
def getHistograms(cols, data):
    f, axes = plt.subplots(4, 3, figsize=(20,20))
    rowIdx = 0
    colIdx = 0
    for col in cols:
        if colIdx > 2:
            rowIdx = rowIdx + 1
            colIdx = 0
            
        if col == 'GarageYrBlt':
            continue
        #print(rowIdx,colIdx)
        ax = sns.histplot(x=col, data=data, ax=axes[rowIdx,colIdx]).set_title(col)
        data[col].value_counts()
        colIdx += 1
    f.tight_layout()
```

# Value Counts of "has" columns

```python
def plotIndicators(data):
    f, axes = plt.subplots(3,3, figsize=(10,10))
    cols = ['Garage', 'Bsmt', 'Alley_Access','Fireplace', 'Pool', 'Fence', 'Misc_Feature', 'MasVnr', 'BeenRemodeled']
    rowIdx = 0
    colIdx = 0
    for idx, col in enumerate(cols):
        if colIdx > 2:
            rowIdx = rowIdx + 1
            colIdx = 0
        
        grouped_garage = data.groupby(f'has{col}').mean().reset_index()
        ax = sns.barplot(grouped_garage[f'has{col}'],grouped_garage['SalePrice'], ax = axes[rowIdx,colIdx])
        colIdx += 1
    f.tight_layout()
```

```python
plotIndicators(train)
```
