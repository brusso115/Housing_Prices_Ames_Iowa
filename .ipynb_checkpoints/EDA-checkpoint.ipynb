```python
train = pd.read_csv('train_filled_na.csv')
test = pd.read_csv('test_filled_na.csv')
train.drop('Id', axis = 1, inplace=True)
```

# Removing Outliers

```python
iso = IsolationForest(contamination=0.1)
yhat = iso.fit_predict(train)
```

```python
mask = yhat != -1
```

```python
train = train.loc[mask, :]
```

```python
X = train.loc[:, train.columns != 'SalePrice']
y = train['SalePrice']
```

# Select K Best

```python
def get_k_best(k, X, y):
    k_best = SelectKBest(f_regression, k=k).fit(X, y)
    X[X.columns[k_best.get_support()]]

    k_best_scores = zip(X.columns, k_best.scores_)
    sorted_coef_importance = sorted(list(k_best_scores),key= lambda x: x[1],reverse=True)
    sorted_coef_importance
    print(sorted_coef_importance)
    f, ax = plt.subplots(figsize=(13, 9))
    sns.barplot([x[0] for x in sorted_coef_importance[:k]], [x[1] for x in sorted_coef_importance[:k]])
    plt.xticks(rotation=90)
    plt.title('Select K Best')
```

```python
get_k_best(30, X=train.loc[:, train.columns != 'SalePrice'], y=train['SalePrice'])
```

# Correlation b/w Variables

```python
non_dummy_cols = [col for col in X if 'Dummy' not in col]
non_has_cols = [col for col in X if 'has' not in col]
custom_features = ['TotalSF', 'BsmtBaths', 'GrLivAreaHighQualitySF', 'TotalIndoorSF','Baths','NumFloors']
X = X[non_dummy_cols]
X = X[non_has_cols]
X = X.drop(custom_features, axis=1)
```

```python
corr = X.corr()
```

```python
def get_redundant_pairs(df):
    '''Get diagonal and lower triangular pairs of correlation matrix'''
    pairs_to_drop = set()
    cols = df.columns
    for i in range(0, df.shape[1]):
        for j in range(0, i+1):
            pairs_to_drop.add((cols[i], cols[j]))
    return pairs_to_drop

def get_top_abs_correlations(df, n=5):
    au_corr = df.corr().abs().unstack()
    labels_to_drop = get_redundant_pairs(df)
    au_corr = au_corr.drop(labels=labels_to_drop).sort_values(ascending=False)
    return au_corr[0:n]


mostCorrelatedFeat = get_top_abs_correlations(corr, 1000)
```

# Value Counts

```python
train_org = pd.read_csv('train.csv')
train_org.drop('Id', axis = 1, inplace=True)
```

```python
def getValueCounts(cols, data):
    f, axes = plt.subplots(9, 5, figsize=(10,10))
    rowIdx = 0
    colIdx = 0
    for col in cols:
        if colIdx > 4:
            rowIdx = rowIdx + 1
            colIdx = 0
        value_counts = data[col].value_counts()
        ax = sns.barplot(x=value_counts.index, y=value_counts, data=data, ax=axes[rowIdx,colIdx]).set_title(col)
        colIdx += 1
        
        
    f.tight_layout()
```

```python
cat_cols = train_org.columns[train_org.dtypes == 'object'] | train_org.columns[train_org.columns == 'MSSubClass']
getValueCounts(cat_cols, train_org)
```

# Boxplots

```python
def getBoxplots(cols, data):
    f, axes = plt.subplots(round(len(cols)/2)+1, 2, figsize=(10,80))
    rowIdx = 0
    colIdx = 0
    for col in cols:
        if colIdx > 1:
            rowIdx = rowIdx + 1
            colIdx = 0
        #print(rowIdx,colIdx)
        if col == 'GarageYrBlt':
            continue
        ax = sns.boxplot(x=col, y='SalePrice', data=data, ax=axes[rowIdx,colIdx]).set_title(col)
        data[col].value_counts()
        colIdx += 1
    f.tight_layout()
```

```python
getBoxplots(cat_cols, train_org)
```

```python
sns.boxplot(x='Neighborhood', y='SalePrice', data=train_org)
plt.xticks(rotation=90)
```

# Histograms

```python
def getHistograms(cols, data):
    f, axes = plt.subplots(4, 3, figsize=(20,20))
    rowIdx = 0
    colIdx = 0
    for col in cols:
        if colIdx > 2:
            rowIdx = rowIdx + 1
            colIdx = 0
            
        if col == 'GarageYrBlt':
            continue
        #print(rowIdx,colIdx)
        ax = sns.histplot(x=col, data=data, ax=axes[rowIdx,colIdx]).set_title(col)
        data[col].value_counts()
        colIdx += 1
    f.tight_layout()
```

```python
getHistograms(['BsmtUnfSF','TotalBsmtSF','LotFrontage','LotArea','BsmtFinSF1','BsmtFinSF1',
              '1stFlrSF','2ndFlrSF','GrLivArea','GarageArea','TotalSF','SalePrice'], train)
```

```python
def getScatterplots(cols, data):
    f, axes = plt.subplots(4, 5, figsize=(20,10))
    rowIdx = 0
    colIdx = 0
    for col in cols:
        if colIdx > 4:
            rowIdx = rowIdx + 1
            colIdx = 0
        if col == 'GarageYrBlt':
            continue
        #print(rowIdx,colIdx)
        ax = sns.scatterplot(x=col, y='SalePrice', data=data, ax=axes[rowIdx,colIdx]).set_title(col)
        colIdx += 1
    f.tight_layout()
```
